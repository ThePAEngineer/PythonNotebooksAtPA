## Part Seven  
That ratio you just calculated has a name. Physicists call it **purity**: of everything you decided to keep, what fraction was actually the thing you wanted?  

There's a second number hiding here too, and it pulls in the opposite direction. **Efficiency** asks: of all the real muons out there in the data, what fraction did your window actually catch? A window that's too narrow throws away good muons. A window that's too wide scoops up junk.
